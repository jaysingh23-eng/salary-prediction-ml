# 📊 Salary Prediction – Exploratory Data Analysis

This notebook walks through the EDA process step-by-step:
1. Load & inspect the dataset
2. Check missing values & data types
3. Univariate analysis (salary distribution)
4. Bivariate analysis (experience, education, industry, remote work)
5. Correlation analysis
6. Key insights summary

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style='whitegrid', palette='muted')
%matplotlib inline

df = pd.read_csv('../data/job_salary_prediction_dataset.csv')
print('Shape:', df.shape)
df.head()
```

```python
# Data types and missing values
print('Data Types:')
print(df.dtypes)
print('\nMissing Values:')
print(df.isnull().sum())
```

```python
# Summary statistics
df.describe()
```

```python
# 1. Salary Distribution
fig, ax = plt.subplots(figsize=(10, 5))
sns.histplot(df['salary'], kde=True, color='#4C72B0', bins=60, ax=ax)
ax.set_title('Salary Distribution', fontsize=15, fontweight='bold')
ax.set_xlabel('Annual Salary (USD)')
plt.tight_layout()
plt.show()
print(f"Mean: ${df['salary'].mean():,.0f} | Median: ${df['salary'].median():,.0f}")
```

```python
# 2. Experience vs Salary
fig, ax = plt.subplots(figsize=(10, 5))
sns.scatterplot(data=df, x='experience_years', y='salary', alpha=0.3, s=10)
sns.regplot(data=df, x='experience_years', y='salary', scatter=False, color='red')
ax.set_title('Experience vs Salary', fontsize=15, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
# 3. Education vs Salary
edu_order = ['High School', 'Diploma', 'Bachelor', 'Master', 'PhD']
fig, ax = plt.subplots(figsize=(10, 5))
sns.boxplot(data=df, x='education_level', y='salary', order=edu_order, palette='Blues')
ax.set_title('Education Level vs Salary', fontsize=15, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
# 4. Top Job Titles by Average Salary
top_jobs = df.groupby('job_title')['salary'].mean().sort_values(ascending=False)
fig, ax = plt.subplots(figsize=(12, 6))
sns.barplot(x=top_jobs.values, y=top_jobs.index, palette='muted')
ax.set_title('Average Salary by Job Title', fontsize=15, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
# 5. Industry Salary Comparison
industry_salary = df.groupby('industry')['salary'].mean().sort_values(ascending=False)
fig, ax = plt.subplots(figsize=(12, 6))
sns.barplot(x=industry_salary.values, y=industry_salary.index, palette='Set2')
ax.set_title('Average Salary by Industry', fontsize=15, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
# 6. Remote Work vs Salary
fig, ax = plt.subplots(figsize=(8, 5))
sns.boxplot(data=df, x='remote_work', y='salary', order=['Yes', 'Hybrid', 'No'], palette='pastel')
ax.set_title('Remote Work Policy vs Salary', fontsize=15, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
# 7. Correlation Heatmap
fig, ax = plt.subplots(figsize=(8, 6))
num_df = df[['experience_years', 'skills_count', 'certifications', 'salary']]
sns.heatmap(num_df.corr(), annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
ax.set_title('Correlation Heatmap', fontsize=15, fontweight='bold')
plt.tight_layout()
plt.show()
```

## 🔑 Key Insights

1. **Experience strongly drives salary** — the regression line shows a clear positive trend
2. **PhD holders earn the most on average** — education has a clear monotonic relationship with pay
3. **AI Engineer and ML Engineer top the salary rankings** — followed by Product Managers
4. **Finance and Technology** pay the highest average salaries across industries
5. **Remote workers earn slightly more** than in-office workers on average
6. **Skills count and certifications** have a moderate positive correlation with salary
7. **Salary distribution is roughly normal** with a slight right skew
